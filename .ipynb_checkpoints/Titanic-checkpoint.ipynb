```python
#Import all dependecies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as ms
%matplotlib inline
```

```python
#read csv file
data=pd.read_csv("titanic.csv")
```

```python
# Preview of the DataFrame
data.head()
```

```python
#Display info about data
data.info()
```

```python
# Display a statistical overview
data.describe()
```

```python
ms.matrix(data)
```

```python
sns.jointplot(x='Fare',y='Age',data=data)
```

```python
sns.distplot(data['Fare'])
```

```python
#Overview correlation
data.corr()
```

```python
# Visualize correlation
sns.heatmap(data.corr(),cmap='coolwarm')
plt.title('data.corr()')
```

```python
sns.swarmplot
sns.swarmplot(x='Pclass',y='Age',data=data,palette='Set1')
```

```python
#Survival count
sns.set_style('whitegrid')
sns.countplot(x='Survived',data=data,palette='RdBu_r')
```

```python
#Survival count based on gender
sns.set_style('whitegrid')
sns.countplot(x='Survived',hue='Sex',data=data,palette='RdBu_r')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Survived',hue='Pclass',data = data,palette='rainbow')
```

```python
data['Fare'].hist(color = 'green', bins = 40, figsize = (8,3))
```
